```python
import pandas as pd
import numpy as np
```

```python
liq = pd.read_csv('https://raw.githubusercontent.com/TrueGRITT38/Soft-Sensor-Data/main/Liquid_Data.csv')
liq = liq.rename(columns = {'Temperature (C)':'rtemp'})
liq.head()
```

```python
vap = pd.read_csv('https://raw.githubusercontent.com/TrueGRITT38/Soft-Sensor-Data/main/Vapor_Data.csv')
vap = vap.rename(columns = {'Temperature (C)':'rtemp'})
vap.head()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/TrueGRITT38/Soft-Sensor-Data/main/HEX_data.csv')
data['rtemp'] = data['TE-800-07/AI1/OUT.CV °C'].round(0)
data.head()
```

```python
data = data.merge(liq[['rtemp','Vol (Vf) (m3/kg)']],how='left',on='rtemp').fillna(0)
data = data.merge(vap[['rtemp','Vol (Vg) (m3/kg)']],how='left',on='rtemp').fillna(0)
data['Spec V'] = 1/data['FIT-800-00/DENSITY/OUT.CV kg/m3']
```

```python
data['Quality'] = (data['Spec V']-data['Vol (Vf) (m3/kg)'])/(data['Vol (Vg) (m3/kg)']-data['Vol (Vf) (m3/kg)'])
data.head()
```
